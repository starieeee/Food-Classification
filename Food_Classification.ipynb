```python
#Choose two foods that want to clarify
labelA = 'samosa'
labelB = 'spring_rolls'
```

- Removing all the images that don't have labelA or labelB in order to handling easier
- Renaming images with labelA or labelB to have that label in their filename

```python
for img in get_image_files(food_path):
    if labelA in str(img):
        img.rename(f"{img.parent}/{labelA}_{img.name}")
    elif labelB in str(img):
        img.rename(f"{img.parent}/{labelB}_{img.name}")
    else:
        os.remove(img)

# Count the total number of image files after renaming or removing
num_images = len(get_image_files(food_path))
print("Total number of image files:", num_images)
```

**Training**

```python
def GetLabel(fileName):
  return fileName.split('_')[0]
  GetLabel("samosa_734186.jpg")
```

```python
dls = ImageDataLoaders.from_name_func(
    food_path, get_image_files(food_path), valid_pct=0.2, seed =420,
    label_func = GetLabel, item_tfms = Resize(200)
)
dls.valid.show_batch()
```

```python
learn = cnn_learner(dls, resnet34, metrics=error_rate, pretrained=True)
learn.fine_tune(epochs=10)
```

Upload our own image

```python
from google.colab import files
uploader = files.upload()

for img in uploader.items():
    uploadedImg = img[0]
    img = PILImage.create(uploadedImg)
    img.show()

    label, _, probs = learn.predict(img)
    print("This is a", label)
    print(f"{labelA} {probs[1].item():.6f}")
    print(f"{labelB} {probs[0].item():.6f}")
```
